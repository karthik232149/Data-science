```python
import matplotlib.pyplot as plt
import pandas as pd
     
     
a = pd.read_csv('ktk.csv')  
print(a)
x=a.iloc[:,:].values
print(x)
 
# print(x)
     
#no traing test required
from sklearn.cluster import KMeans
kmeans=KMeans(init='k-means++',n_clusters=2,random_state=0).fit(x)
p=kmeans.predict(x)
print(p)
     
     
colors=list(map(lambda x:'#fcc603' if x==1 else '#03dffc',p))
plt.scatter(x[:,0],x[:,1],c=colors,marker='*',picker=True)
     
plt.show()

```
